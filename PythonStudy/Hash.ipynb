Hashmaps: Basically something that has a key and a definitive outcome. 
Look for a key

Use a hash function on an array to create a hash map
if index already exists it creates a new mutation

Keys are immutable, you can not change the key after first initialized

Instead of using an array, you could use a tuple if you wanted to use it like that.

```python
city_map = {}

cities = ["calgary", "Vancouver", "toronto"]
city_map["Canada"] = []
city_map["Canada"] += cities
```

```python
city_map.keys()
```

```python
city_map.values()
```

```python
city_map.items()
```

You might encounter problems appending because first you would need to make a key then initialize values, in order to get around this, you can use defaultdicts
Defaultdicts make it so that when you try to access a key that does not exist, it will just create a key with that value. 

```python
from collections import defaultdict
city_map = defaultdict(list)
```
